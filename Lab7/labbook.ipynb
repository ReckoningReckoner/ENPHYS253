External oscillating voltage causes resonant behaviour in some electrical circuits, the simplest example is shown in figure 1 below.

-The circuit in Figure 2 was assembled
-The Variable capacitor was set to 0.005 microFarads
-The left input channel of the Agilent 54621A (60MHz) oscilliscope was connected across the resistor R1
-The voltage is directly dependent on current by

$V_{R}=R_{1}I$

-The voltage was read from the oscilliscope as a function of frequency $f$ generated by BK Precision 4011

-The resonant frequency $f_{0}$ was recorded along with the respective capacitance varied over a range of 1-10 nanoFarads in increments of 1. The Capacitance was generated by a type 1419-A Polydtyrene Decade Capacitor with an uncertainty of 0.5%.

<img src="circuit.png">


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sci
Part1Resonance = pd.read_excel('Part1Resonance.xlsx')
print("Table 1: Resonant frequency as a function of Capacitance")
Part1Resonance
```

-The eqution below was used to plot the inverse of the square frequency as a function of Capacitance. L and $C_{s}$ were deduced from the graph

$1/f_{0}^2 = 4\pi^2L(C+C_{s})$

```python
from scipy import stats
import numpy as np
Part1Resonance1 = Part1Resonance.as_matrix()
resonance_frequency = Part1Resonance1[:,0]*1000
Capacitance = Part1Resonance1[:,1]
slope, intercept, r_value, p_value, std_err = stats.linregress(Capacitance*1e-6,1/resonance_frequency**2 )
print('The Value for the inductor is {:.3f}'.format(slope))
print('The Value for Cs is {:.11f}'.format(intercept))

plt.plot(Capacitance*1e-6,1/resonance_frequency**2 ,'.')
plt.ylabel("Capacitance of Polydtyrene Decade Capacitor [Farads]")
plt.xlabel("The inverse of the square of resonance frequency")
plt.show()


```
